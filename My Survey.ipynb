```python
facebook_using = data[data.pform=="Facebook"]
```

```python
y = facebook_using.use_begin.sort_values(ascending=True)
new_y =list(dict.fromkeys(y))
x= list(facebook_using.use_begin)
my_dict = [x.count(i) for i in new_y]
```

```python
print(new_y)
print(my_dict)
```

```python
data_example = pd.DataFrame({"year":new_y,"number":my_dict})
```

```python
data_example.plot("year","number")
plt.title("How the number of Facebook users increased")
plt.xlabel("Years")
plt.ylabel("Users Numbers")
plt.grid(True)
```

```python
study_kind= data[data.using_kind=="Communication"]

ss =study_kind.age.sort_values(ascending=True)
new_study =list(dict.fromkeys(ss))
print(new_study)
new_xstudy= list(study_kind.age)
my_dict_study = [new_xstudy.count(i) for i in new_study]
my_dict_study
```

```python
calc_lat = data[data.city=="LAT"]
cc =calc_lat.max_hours.sort_values(ascending=True)
new_calc =list(dict.fromkeys(cc))
print(new_calc)
new_xcalc= list(calc_lat.max_hours)
lat_max_hours = [new_xcalc.count(i) for i in new_calc]
lat_max_hours
```
