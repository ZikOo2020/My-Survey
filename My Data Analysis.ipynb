```python
import pandas as pd 
import math as mt

data = pd.read_csv("C:/Users/Zualfiqar/Desktop/Social Media.csv")
```

```python
data = data.drop("country",axis=1)
```

```python
data = data.drop("id",axis=1)
```

```python
data = data.drop("rate_2",axis=1)
```

```python
data.head()
```

```python
data[data.duplicated(keep=False)]
```

```python
data[data.isna().any(axis=1)]
```

```python
min_hours = data.min_hours.fillna(mt.floor(data.min_hours.min()))
data.min_hours=min_hours
data[data.isna().any(axis=1)]
```

```python
max_hours = data.max_hours.fillna(mt.ceil(data.max_hours.mean()))
data.max_hours = max_hours
data[data.isna().any(axis=1)]
```

```python
min_person = data.min_person.fillna(mt.floor(data.min_person.min()))
data.min_person = min_person
data[data.isna().any(axis=1)]
```

```python
max_person = data.max_person.fillna(mt.ceil(data.max_person.mean()))
data.max_person = max_person
data[data.isna().any(axis=1)]
```

```python
count =0
for i in range(0,len(data.using_kind.isna())):
    if data.using_kind[i]!="NEWS" and data.using_kind[i]!="Studying" and data.using_kind[i]!="Communication" and data.using_kind[i]!="Work":
        count = count+1
        if count <= 6:
            data.using_kind[i]="Studying"
        else:
            data.using_kind[i]="Work"
count
data[data.isna().any(axis=1)]
```

```python
facebook_using = data[data.pform=="Facebook"]
y = facebook_using.use_begin.sort_values(ascending=True)
new_y =list(dict.fromkeys(y))
x= list(facebook_using.use_begin)
my_dict = [x.count(i) for i in new_y]
```

```python
data_example = pd.DataFrame({"year":new_y,"number":my_dict})
data_example.plot("year","number",figsize=(8,6))
plt.title("How the number of Facebook users increased")
plt.xlabel=("Years")
plt.ylabel=("Users Numbers")
plt.grid(True)
```

```python
#Studying the users from latakia city
latakia_users = data[data.city=="Latakia"]

import numpy as np
from matplotlib.ticker import PercentFormatter

plt.hist(latakia_users.opinion,weights=np.ones(len(latakia_users.opinion)) / len(latakia_users.opinion))
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.grid(True)
plt.figsize=(15,10)
plt.title("The Impact of Socail Media on the users from Latakia")
plt.show()
```

```python
plt.hist(latakia_users.rate,weights=np.ones(len(latakia_users.rate)) / len(latakia_users.rate))
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.grid(True)
plt.figsize=(15,10)
plt.title("Rating the Socail Media by the users from Latakia")
plt.show()
```

```python
plt.hist(latakia_users.max_hours,weights=np.ones(len(latakia_users.max_hours)) / len(latakia_users.max_hours))
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.grid(True)
plt.figsize=(15,10)
plt.title("Hours which the users from latakia spend it in the day on Social Media")
plt.show()
```
